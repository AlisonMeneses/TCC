```python
import numpy as np
import pandas as pd
```

# Leitura dos dados

```python
df = pd.read_csv("GroceryProductsPurchase.csv", header = None)
df.head()
```

# Removendo a Primeira Linha do DataFrame

```python
df = df.drop(0)
df.head()
```

# Trocando os NAN por Zeros

```python
df = df.replace(np.nan, 0)
df.head()
```

# AnÃ¡lise dos Dados

```python
df.describe()
```

# Convertendo o DataFrame em Uma Lista de Listas para Usar os Tamanhos VariÃ¡veis das Compras Removendo os Zeros

```python
def removeZerosInList(_list):
    return list(filter(lambda x: x != 0, _list))
```

```python
list_all_transactions = []

for index, row in df.iterrows():
    list_single_transaction = row.values.tolist()
    list_single_transaction = removeZerosInList(list_single_transaction)
    list_all_transactions.append(list_single_transaction)
```

```python
for index in range(10):
    print(list_all_transactions[index])
```

# Pre-processamento dos dados

## Criando um NumPy array de one-hot encoded pois Ã© o formato que o apriori usa como entrada

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(list_all_transactions).transform(list_all_transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

df.head()
```

# Utilizando o Algoritmo Apriori

## Calculando os Suportes dos Itens

```python
from mlxtend.frequent_patterns import apriori

frequent_items = apriori(df, min_support = 0.01, use_colnames = True)
frequent_items.sort_values(by=['support'], ascending = False).head(10)
```

## Estabelecendo Regras de AssociaÃ§Ã£o

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)
```

## Ordenando pelo Lift

```python
rules = rules.sort_values(by=['lift'], ascending = False).drop(['antecedent support', 'consequent support', 'leverage', 'conviction'], axis=1)
rules.head(10)
```
